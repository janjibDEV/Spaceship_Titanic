# Dataset Information
* link: https://www.kaggle.com/competitions/spaceship-titanic/data

```python
# import necessary libraries for data cleaning
import pandas as pd
import numpy as np
```

```python
# read train.csv
df = pd.read_csv('train.csv')
# display first 5 row
df.head()
```

```python
# View general info of the dataframe
df.shape,df.info(),df.describe()
```

```python
# passenger id
'''
PassengerId - A unique Id for each passenger. Each Id takes the form gggg_pp where gggg indicates a group the passenger is travelling with and pp is their number within the group. 
People in a group are often family members, but not always.
'''
# Create new columns: Group and Pnum
df['Group'] = df['PassengerId'].apply(lambda x: int(str(x)[:4]))
df.head()
```

```python
# delete PassengerId
del df['PassengerId']
df.head()
```

```python
# HomePlanet
'''
HomePlanet - The planet the passenger departed from, typically their planet of permanent residence.
'''
q = pd.Categorical(df['HomePlanet'])
q.describe()
```

```python
# set NaN value to Earth
df['HomePlanet'] = df['HomePlanet'].fillna('Earth')
q = pd.Categorical(df['HomePlanet'])
q.describe()
```

```python
# CryoSleep
'''
CryoSleep - Indicates whether the passenger elected to be put into suspended animation for the duration of the voyage. 
Passengers in cryosleep are confined to their cabins.
'''
q = pd.Categorical(df['CryoSleep'])
q.describe()
```

```python
# set NaN values to False
df['CryoSleep'] = df['CryoSleep'].fillna(False)
q = pd.Categorical(df['CryoSleep'])
q.describe()
```

```python
df['CryoSleep'] = df['CryoSleep'].apply(lambda x: int(x))
df['CryoSleep']
```

```python
# Cabin
#Cabin - The cabin number where the passenger is staying. 
#Takes the form deck/num/side, where side can be either P for Port or S for Starboard.
test = np.where(df['Cabin'].isnull())
len(test[0])
# 199 values are missing
```

```python
# create column deck
q = df['Cabin'].apply(lambda x: str(x)[0])
c = pd.Categorical(q)
c.describe()
```

```python
q = np.where(q == 'n','T',q)
c = pd.Categorical(q)
c.describe()
```

```python
df['Deck'] = q
df['Deck']
```

```python
# create column Cnum
q = df['Cabin'].apply(lambda x: str(x)[2])
c = pd.Categorical(q)

q = np.where(q == 'n','0',q)
c = pd.Categorical(q)

q = np.array(list(map(int,q)))
df['Cnum'] = q
```

```python
# create column port
q = df['Cabin'].apply(lambda x: str(x)[-1])
c = pd.Categorical(q)

q = np.where(q == 'n','S',q)
c = pd.Categorical(q)

df['Port'] = q
df.head()
```

```python
# delete cabin column
del df['Cabin']
```

```python
q = pd.Categorical(df['VIP'])
q.describe()

df['VIP'] = df['VIP'].fillna(False)
df['VIP'] = df['VIP'].apply(lambda x: int(x))
df['VIP']
```

```python
# replace null values with 0.0 for RoomService, FoodCourt, ShoppingMall, Spa, VRDeck
df['RoomService'] = df['RoomService'].fillna(0)
df['FoodCourt'] = df['FoodCourt'].fillna(0)
df['ShoppingMall'] = df['ShoppingMall'].fillna(0)
df['Spa'] = df['Spa'].fillna(0)
df['VRDeck'] = df['VRDeck'].fillna(0)
```

```python
# delete name column
del df['Name']
```

```python
# separate label
y = df['Transported']
del df['Transported']
```

```python
df
```

```python
cleaned_df = df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(cleaned_df,y,random_state = 42)
```

```python
# Hot encoding 'HomePlanet','Destination','Deck',Port

enc_column = ['HomePlanet','Destination','Deck','Port']

from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder(handle_unknown='ignore',sparse=False)
OH_cols_train = pd.DataFrame(enc.fit_transform(X_train[enc_column]))
OH_cols_valid = pd.DataFrame(enc.transform(X_valid[enc_column]))

OH_cols_train.index = X_train.index
OH_cols_valid.index = X_valid.index

num_x_train = X_train.drop(enc_column,axis=1)
num_x_valid = X_valid.drop(enc_column,axis=1)

OH_X_train = pd.concat([num_x_train,OH_cols_train],axis=1)
OH_X_valid = pd.concat([num_x_valid,OH_cols_valid],axis=1)

OH_X_train
```

```python
import math
# fill null values in age with mean value
mean_age = math.floor(OH_X_train['Age'].mean())
OH_X_train['Age'] = OH_X_train['Age'].fillna(mean_age)
OH_X_valid['Age'] = OH_X_valid['Age'].fillna(mean_age)
```

```python
OH_X_train
```

```python
# scale 'RoomService',	'FoodCourt',	'ShoppingMall',	'Spa',	'VRDeck' column
from sklearn.preprocessing import StandardScaler

col_names = ['RoomService',	'FoodCourt',	'ShoppingMall',	'Spa',	'VRDeck']

scaled_train = OH_X_train.copy()
scaled_valid = OH_X_valid.copy()

features_train = scaled_train[col_names]
features_valid = scaled_valid[col_names]

scaler = StandardScaler().fit(features_train.values)
features_train = scaler.transform(features_train.values)
features_valid = scaler.transform(features_valid.values)

scaled_train[col_names] = features_train
scaled_valid[col_names] = features_valid

scaled_train
```

```python
# build a ml model using xgb and grid search

import xgboost as xgb
from sklearn.model_selection import GridSearchCV

estimator = xgb.XGBClassifier(
    objective= 'binary:logistic',
    nthread=4,
    seed=42
)

parameters = {
    'max_depth': range (2, 10, 1),
    'n_estimators': range(60, 220, 40),
    'learning_rate': [0.1, 0.01, 0.05, 0.2, 0.3, 0.4]
}

grid_search = GridSearchCV(
    estimator=estimator,
    param_grid=parameters,
    scoring = 'roc_auc',
    n_jobs = 10,
    cv = 10,
    verbose=True
)

grid_search.fit(scaled_train,y_train)

```

```python
grid_search.best_estimator_
```

```python
xgb_cl = xgb.XGBClassifier(max_depth=4, n_estimators=180, nthread=4, seed=42)

xgb_cl.fit(scaled_train,y_train)

preds = xgb_cl.predict(scaled_valid)
```

```python
# check accuracy
from sklearn.metrics import accuracy_score

ac = accuracy_score(y_valid,preds)
ac
```
